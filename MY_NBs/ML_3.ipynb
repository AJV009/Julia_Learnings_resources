```julia
using JuliaAcademyData; activate("Foundations of machine learning")
```

```julia
using Plots; gr()
using Images, Statistics

sig(x,w,b) = 1 / (1 + exp(w*x + b))

apple = load(datapath("data/10_100.jpg"))
banana = load(datapath("data/104_100.jpg"))

apple_gr_a = mean(Float64.(green.(apple)))
banana_gr_a = mean(Float64.(green.(banana)))

w = 20
b = 12.5

plot(x -> sig(x,w,b), 0, 1, label="model", legend=:topleft, lw=3)
scatter!([apple_gr_a],[0.0],label="Apple")
scatter!([banana_gr_a],[0.0],label="Banana")
```

```julia
plotly()
```

```julia
L(w,b) = (0 - sig(apple_gr_a,w,b))^2 + (1-sig(banana_gr_a,w,b))^2

w_ran = 10:0.1:13
b_ran = 0:0.1:20

L_val = [L(w,b) for b in b_ran, w in w_ran]

w = 11.5
b = 10
p1 = surface(w_ran, b_ran, L_val, xlabel="w", ylabel="b", cam=(70,40), cbar=false, leg=false)
# p1 = contour(w_ran,b_ran,L_val, levels=0.05:0.1:1, xlabel="w", ylabel="b", cam=(70,40), cbar=false, leg=false)
scatter!(p1, [w], [b], markersize=5, color = :blue)

p2 = plot(x->sig(x,w,b), 0, 1, label="Model", legend = :topleft, lw=3)
scatter!(p2, [apple_gr_a],  [0.0], label="Apple", markersize=10)
scatter!(p2, [banana_gr_a], [1.0], label="Banana", markersize=10, xlim=(0,1), ylim=(0,1))
plot(p1, p2, layout=(2,1))
```
