```julia
using Images, Statistics

apple = load(datapath("data/10_100.jpg"))
banana = load(datapath("data/104_100.jpg"))

apple_green_amount = mean(Float64.(green.(apple)))
banana_green_amount = mean(Float64.(green.(banana)))
println("Green of Apple = $apple_green_amount")
println("Green of Banana = $banana_green_amount")
```

```julia
sig(x, w, b) = 1 / (1 + exp(-w * x + b))
```

```julia
using Plots; gr()
```

```julia
w = 20
b = 12.5
# w = 25.58; b = 15.6
plot(x -> sig(x, w, b), xlim=(-0,1), ylim=(-0.1,1.1), label="model", legend=:topleft, lw=3)

scatter!([apple_green_amount], [0.0], label="apple", ms=5)
scatter!([banana_green_amount], [1.0], label="banana", ms=5)
```
